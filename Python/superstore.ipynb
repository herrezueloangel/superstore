```python
import pandas as pd

df = pd.read_csv("Sample - Superstore.csv",sep=',', encoding='latin1')

#El formato de estas dos columnas están erroneas, las paso a formate Datetime:
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])

#Para hacer cálculos futuros separamos las fechas por mes y por año
df["Order Month Name"] = df["Order Date"].dt.month_name()
df["Order Year"] = df["Order Date"].dt.year
```

```python
#Vemos la suma de todos los años para analizar su evolcuión, si hubo crecimiento, si hubo algún año diferente...
df.groupby("Order Year")["Sales"].sum().round(2)
```

```python
#Verificamos si todos los años tienen todos los meses para ver si realmente todos los años con comparables.
df.groupby("Order Year")["Order Month Name"].nunique()
```

```python
#Regiones por ventas y beneficios
df.groupby("Region")[["Sales", "Profit"]].sum().sort_values(by="Sales", ascending=False)

#Creamos una nueva columna con el beneficio porcentual para ver más exacto a que equivale ese beneficio con respecto a las ventas
df["Profit Margin"] = round((df['Profit']/df['Sales'])*100,2)
#Visualización de ventas vs cantidad beneficio vs % beneficio
df[["Sales", "Profit", "Profit Margin"]].head()
```

```python
#Agrupo por descuentos para ver como son los beneficios depende de los descuentos aplicados
df.groupby("Discount")[["Sales", "Profit", "Profit Margin"]].mean()
```

```python
#Calculamos la media de las tres columnas y agrupamos por la categoria para ver la media del margen de beneficio
df.groupby("Category")[["Sales", "Profit", "Profit Margin"]].mean()
```

```python
#Ahora vemos la categoria y su descuento para ver que margen de beneficio tiene
df.groupby(["Category", "Discount"])["Profit Margin"].mean().reset_index()
```

```python
#Top 10 ciudades

df.groupby('City')['Sales'].sum().sort_values(ascending=False).head(10)
```

```python
#Ventas por año
df['Year'] = df['Order Date'].dt.year
df.groupby('Year')['Sales'].sum()
```

```python
#Top clientes
df.groupby('Customer Name')['Sales'].sum().sort_values(ascending=False).head(10)
```
